# Customer Sales Insights Analysis  
## Retail Customer Segmentation & Purchase Behaviour Analytics

This notebook focuses on analysing the final cleaned and merged retail dataset. The goal is to identify customer segment patterns, product performance, pack-size preferences, and sales behaviour using Pandas and NumPy.

The analysis in this notebook provides initial business insights before moving into SQL analysis and Power BI dashboard development.

```python
# importing numpy library

import numpy as np
```

```python
# importing pandas library

import pandas as pd
```

```python
# loading merged dataset 

final_df = pd.read_csv(
    r"..data/Processed_data/final_customer_sales_data.csv"
)

final_df.head()
```

```python
# checking shape of the data 

final_df.shape
```

```python
# a statistical summary of this data

final_df.describe()
```

```python
# 1. Sales by lifestage

revenue_by_lifestage = final_df.groupby("LIFESTAGE")["TOT_SALES"].sum().sort_values(ascending=False)
print(revenue_by_lifestage)
```

It shows that older singles/ couples are generating the highest revenue while new families are generating lowest revenue

```python
# 2. which product generated highest revenue and least revenue 

prod_revenue = final_df.groupby("PROD_NAME")["TOT_SALES"].sum().sort_values(ascending=False)
print("top 3 :\n",prod_revenue.head(3))
print("bottom 3 :\n",prod_revenue.tail(3))
```

Dorito Corn Chp Supreme generated highest revenue while Woolworths Medium Salsa generated least revenue

```python
# 3. which product was highly consumed and least consumed

prod_consumed = final_df.groupby("PROD_NAME")["PROD_QTY"].sum().sort_values(ascending=False)
print("top 3 :\n",prod_consumed.head(3))
print("bottom 3 :\n",prod_consumed.tail(3))
```

Kettle Mozzarella Basil & Pesto was highly consumed and French Fries Potato Chips was least consumed 

```python
# changing dataype of date 

final_df["DATE"] = pd.to_datetime(final_df["DATE"])

# checking type
final_df.dtypes
```

```python
# 4. which month generated highest sales

m_sales = final_df.groupby(pd.Grouper(key="DATE",freq="ME"))["TOT_SALES"].sum().reset_index(name="Total sales").sort_values("Total sales",ascending=False)
m_sales["DATE"] = m_sales["DATE"].dt.strftime("%B %Y")
print(m_sales)
```

December 2018 generated the highest revenue while February 2019 generated lowest revenue 

```python
# 5. finding high value customers

hv_customers = final_df.groupby("LYLTY_CARD_NBR")["TOT_SALES"].sum().sort_values( ascending = False )
print(hv_customers.head(5))
```

High-value customers were identified by grouping transactions by `LYLTY_CARD_NBR` and calculating total customer spend. Customers with the highest total spend represent the most valuable loyalty customers in the dataset.

```python
# 6. categorising dataset based on sales 

final_df["Sales_Category"] = np.where(
    final_df["TOT_SALES"] >= final_df["TOT_SALES"].median(),
    "High Sale",
    "Low Sale"
)

final_df
```

```python
# 7. categorising products based on their pack size

final_df["PACK_CATG"] = np.select  (
    [
        final_df["PACK_SIZE_(g)"] <= 150 ,
        final_df["PACK_SIZE_(g)"].between(160,200),
        final_df["PACK_SIZE_(g)"] > 200
    ],
    [
        "Small pack",
        "Medium pack",
        "Large pack"
    ],
    default = "unknown"
)

final_df
```

## Key Findings Summary

The initial Pandas and NumPy analysis highlighted several important sales and customer behaviour patterns:

- Customer segments show different levels of revenue contribution, with some lifestage and premium customer groups generating higher total sales than others.
- Product revenue and product quantity results show that the highest revenue-generating products are not always the same as the most purchased products by quantity.
- Monthly sales analysis identified the strongest revenue-generating month within the available transaction period.
- High-value customers were identified by grouping transactions by `LYLTY_CARD_NBR` and calculating total customer spend.
- NumPy-based feature engineering was used to create sales and pack-size categories, supporting deeper customer and product segmentation.

These findings provide a strong foundation for the next stage of the project, where SQL will be used for structured business analysis and Power BI will be used for dashboard storytelling.
